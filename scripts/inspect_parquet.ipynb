# Parquet File Inspector
A minimal notebook for quick inspection of Parquet files.

## 1 — Imports  
Load `pandas` and `pyarrow`, and set display options.


```python
import pandas as pd
import pyarrow.parquet as pq  # required by pandas.read_parquet

pd.set_option("display.max_columns", None)  # show all columns in outputs
```

## 2 — Specify the Parquet file path  
Edit the `file_path` variable in the next cell to point to the file you want to inspect.


```python
# === User parameter ===
file_path = "path/to/your/file.parquet"  # <-- change this
```

## 3 — Load the Parquet file and preview  
Read the file into a DataFrame, print its shape, and show the first few rows.


```python
df = pd.read_parquet(file_path)
print(f"Loaded {len(df):,} rows × {df.shape[1]} columns")
df.head()
```

## 4 — Data types and non-null counts  
Display column dtypes, non-null counts, and estimated memory usage.


```python
print("Data types and non-null counts:")
df.info(memory_usage="deep")
```

## 5 — Basic statistics  
Generate descriptive statistics for numeric **and** categorical columns.


```python
# Numeric and categorical summary
stats = df.describe(include="all").T
stats
```

## 6 — Missing-value audit  
Show how many missing values each column contains, sorted from most to least.


```python
print("Missing values per column:")
df.isna().sum().sort_values(ascending=False)
```
